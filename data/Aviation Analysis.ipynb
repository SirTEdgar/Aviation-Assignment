```python
# Set visualization aesthetics
%matplotlib inline
sns.set_theme(style="whitegrid")

# Load the intermediate clean dataset from Part One
df_final = pd.read_csv("AviationData_Cleaned_Final.csv")

print("--- DATASET LOADING VERIFICATION ---")
print(f"Total Rows Loaded: {df_final.shape[0]}")
print(f"Total Columns Loaded: {df_final.shape[1]}")
print("\nColumns available for Exploratory Data Analysis:")
print(df_final.columns.tolist())

print("\n--- SAFETY METRICS SNAPSHOT ---")
display(df_final[['Severe.Injury.Rate', 'Is.Destroyed']].describe())
```

```python
# Ensure dataframes are explicitly created right here to prevent NameErrors
df_small = df_final[df_final['Estimated.Total.Passengers'] <= 20].copy()
df_large = df_final[df_final['Estimated.Total.Passengers'] > 20].copy()

# Isolate the top 10 lowest risk makes for distributions
small_top10_makes = df_small.groupby('Make_Cleaned')['Severe.Injury.Rate'].mean().nsmallest(10).index
large_top10_makes = df_large.groupby('Make_Cleaned')['Severe.Injury.Rate'].mean().nsmallest(10).index

#Distribution Plot for Small Airplane Makes (Violin Plot)
plt.figure(figsize=(12, 5))
sns.violinplot(
    data=df_small[df_small['Make_Cleaned'].isin(small_top10_makes)],
    x='Make_Cleaned', y='Severe.Injury.Rate', order=small_top10_makes, palette='Blues'
)
plt.title('Distribution of Severe Injury Rates for Top 10 Small Airplane Makes')
plt.xlabel('Manufacturer')
plt.ylabel('Severe Injury Rate')
plt.xticks(rotation=45)
plt.show()

# Distribution Plot for Large Airplane Makes (Strip Plot)
plt.figure(figsize=(12, 5))
sns.stripplot(
    data=df_large[df_large['Make_Cleaned'].isin(large_top10_makes)],
    x='Make_Cleaned', y='Severe.Injury.Rate', order=large_top10_makes, 
    jitter=0.2, size=6, palette='Oranges', alpha=0.7
)
plt.title('Distribution of Severe Injury Rates for Top 10 Large Airplane Makes')
plt.xlabel('Manufacturer')
plt.ylabel('Severe Injury Rate')
plt.xticks(rotation=45)
plt.show()

# Evaluate rate of aircraft destruction by make (Keep lowest 15)
small_destroyed = df_small.groupby('Make_Cleaned')['Is.Destroyed'].mean().nsmallest(15).reset_index()
large_destroyed = df_large.groupby('Make_Cleaned')['Is.Destroyed'].mean().nsmallest(15).reset_index()

print("--- TOP 15 SMALL MAKES: LOWEST HULL DESTRUCTION RATE ---")
print(small_destroyed.to_string(index=False))

print("\n--- TOP 15 LARGE MAKES: LOWEST HULL DESTRUCTION RATE ---")
print(large_destroyed.to_string(index=False))
```

```python
# Occurrences of each unique airplane model/type in the small plane fleet
small_model_counts = df_small['Unique_Plane_Type'].value_counts()

# Filter for small models that have AT LEAST 10 individual incident examples
valid_small_models = small_model_counts[small_model_counts >= 10].index
df_small_filtered_models = df_small[df_small['Unique_Plane_Type'].isin(valid_small_models)]

top10_small_models = (
    df_small_filtered_models.groupby('Unique_Plane_Type')['Severe.Injury.Rate']
    .mean()
    .nsmallest(10)
    .index
)

# Filter data
df_small_plots = df_small_filtered_models[df_small_filtered_models['Unique_Plane_Type'].isin(top10_small_models)]

# Mean Fatal/Seriously Injured Fraction (Bar Plot)
plt.figure(figsize=(12, 5))
sns.barplot(
    data=df_small_plots,
    x='Severe.Injury.Rate',
    y='Unique_Plane_Type',
    order=top10_small_models,
     hue='Unique_Plane_Type',
    legend=False,
    errorbar=None,
    palette='Blues_r'
)
plt.title('Top 10 Small Airplane Models by Lowest Mean Severe Injury Rate\n(Minimum 10 Incidents Required)', fontsize=14, fontweight='bold')
plt.xlabel('Mean Severe Injury Rate', fontsize=12)
plt.ylabel('Airplane Model', fontsize=12)
plt.tight_layout()
plt.show()

# Distributional Plot of Choice (Violin Plot for Small Planes)
plt.figure(figsize=(12, 6))
sns.violinplot(
    data=df_small_plots,
    x='Unique_Plane_Type',
     y='Severe.Injury.Rate',
    order=top10_small_models,
    hue='Unique_Plane_Type',
    legend=False,
    palette='Blues'
)
plt.title('Distribution of Severe Injury Rates Across Top Small Airplane Models', fontsize=14, fontweight='bold')
plt.xlabel('Airplane Model', fontsize=12)
plt.ylabel('Fraction of Passengers Seriously/Fatally Injured', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
